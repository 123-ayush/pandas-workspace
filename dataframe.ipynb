DataFrame

```python
import pandas as pd

# creating df type 1:

info = {
    "Name": ["Ayush", "Arun", "Sunny"],
    "Age": [20, 21, 20]
}

df = pd.DataFrame(info)

print(df, type(df))

print(df.index)
print(df.columns)
```

```python
# creating dataframe type 2:

df = pd.DataFrame([["Ayush", 20], ["Arun", 21], ["Sunny", 20]], columns=["Name", "Age"])

print(df, type(df))

print(df.index)
print(df.columns)
```

```python
# creating dataframe type 3:

import numpy as np

np_arr = np.array([[1,2,3], [4,5,6]])

df = pd.DataFrame(np_arr, columns=["A", "B", "C"])

print(df)
```

Pandas DataFrame Methods
- df.head() -> starting 5 rows, we can also n rows by sending n as parameter
- df.tail() -> last 5 rows
- df.sample() -> any 1 random row
- df.info() -> give summary of data
- df.shape -> return tuple of shape of data
- df.describe() -> give numerical result of data like mean median etc
- df.columns -> give all cols name
- df.nunique() -> give no. of unique rows
