```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4,5])

print(s)
print(type(s))

# indexing 

print(s[3])
```

Indexing two types:
- label based -> prefered to use loc function
- index based -> prefered to use iloc function

```python
#assigning labels to series

s1 = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])

print(s1['a']) # label based
print(s[0]) # index based 

print(s1.loc['a']) # label based using loc
print(s.iloc[0]) # index based using iloc
```

Properties of series:
- Homogenous
- Vectorized Operations
- Handle missing value with NaN
- Mutable values, immutable size

```python
# Homogenous
s2 = pd.Series([1,2,3,4,5,'abc'], index=['a','b','c','d','e',25])
# print(s2)
# print(type(s2))
# print(s2.index)

# vectorized operations
s3 = pd.Series([1, 2, 3, 4, 5])
s4 = pd.Series([10, 20, 30, 40, 50])
print(s3+s4)

# mutable values
s3[0] = 100
print(s3)

# drop function

changed_s3 = s3.drop(0) # due to immutable size we have to create a new series
print(s3)
print(changed_s3)
```
