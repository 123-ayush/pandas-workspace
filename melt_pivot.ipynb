Melt & Pivot
- Melt - Wide to Long format
- melt( id_vars, value_vars, var_name, value_name )
- id_vars -> columns we want to keep 
- value_vars -> we want to convert them in single col
- var_name -> name of new col which store names of vars
- value_name -> name of new col which store values

```python
import pandas as pd
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 18, 22]
})

df
```

```python
melted_df = df.melt(
    id_vars=["country", "year"],
    value_vars=["sales", "profit"],
    var_name="metrics",
    value_name="value"
)
```


- Pivot - Long to Wide format
- pivot( index, columns, values )
- index -> values assigned to each row
- columns -> var_name in melted
- values -> value_name in melted

```python
original = melted_df.pivot(
    index=["country", "year"],
    columns="metrics",
    values="value"    
)
```
