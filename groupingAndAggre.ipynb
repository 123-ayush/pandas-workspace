Grouping & Aggregation of Data
- groupby( )
- mean/ sum / max / min / count etc.
- agg( )

```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_data.csv")
df = df.drop(columns="Unnamed: 0")
df
```

```python
df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()

df.groupby("country")["income"].agg(avg_salary = "mean", median_salary= "median", min_salary = "min", max_salary = "max")
```

```python
df.groupby("country").agg({
    "income": "mean",
    "age": "mean"
})

df.groupby("country").agg(
    avg_income = ("income", "mean"),
    avg_age = ("age", "mean")
)
```
