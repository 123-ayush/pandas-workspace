Transforming Data (Feature Engineering)
- apply( )
- map( )
- assign( )
- replace( old, new )

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df2 = df.copy()

df2["tax"] = df2["income"].apply(lambda x : "20%" if x>=60000 else "10%")

df2
```

Higher Order Functions are function which take another fucntion as parameter.

```python
gender_map = {"Male": "M", "Female": "F", "Unknown": "U"}
df2["gender"] = df2["gender"].map(gender_map)

df2
```

```python
df2 = df2.assign(new_income = df2["income"] * 1.1)
df2
```

```python
df2["country"] = df2["country"].replace("USA","US")
df2
```

Transforming Data
- rename( columns={old: new} )
- sort_values( )
- sort_index( )
- reset_index( )
- rank( )
- Reorder columns

```python
df2.rename(columns={"income":"salary"})
df2.rename(index={1:"first"})
```

```python
df2.sort_values("income")
df2.sort_values("income", ascending=False)

sorted_df = df2.sort_values(["income","age"])

sorted_df.sort_index()

sorted_df

sorted_df.reset_index() # preserve index values
sorted_df.reset_index(drop=True)
```

```python
sorted_df["ranking"] = sorted_df["income"].rank(ascending=False, method="dense")
sorted_df
```

```python
# reorder columns

sorted_df[['id', 'name', 'age', 'country', 'gender', 'income', 'tax', 'ranking', 'new_income']]
```
