Pandas
query() rules
- Expression as a string
- Column based reference
- Backticks for col_name with space or special chars
- Operators (&, |, ~, <=, >=, ==, <, >, etc.)
- @ to reference Python vars
- Chained comparisons

- query return copy of data

```python
aqi_val = 100
df.query("aqi > @aqi_val & temperature > 30")
```

Cleaning Data
- Handle Missing Values
- isna( )/isnull( )
- isnull( ).sum()
- dropna( )
- fillna(value)
- ffill( )
- bfill()

```python
df = pd.read_csv("raw_data.csv")
df.isnull()
df.isna()
```

```python
df.isnull().sum()
df.describe().mean

df.dropna() # drop row with empty cell
df.dropna(axis=1)

cleaned_data = df.copy()
cleaned_data[["id", "age", "income"]].fillna(cleaned_data[["id", "age", "income"]].mean())

df.ffill()
df.bfill()
```
